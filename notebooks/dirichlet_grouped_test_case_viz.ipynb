```python
import os
import json
from pathlib import Path
from collections import defaultdict

import numpy as np
import matplotlib.pyplot as plt
import yaml

def extract_case_label(case_name):
    primary = case_name.split('--')[0]
    primary = primary.split('_seeds')[0]
    return primary

metric_name = 'PBL Mean'
target_cases = ['case14', 'case30', 'case57', 'case118', 'case500', 'case14_30_57', 'case118_500']

results = defaultdict(lambda: defaultdict(list))
model_labels = {}

for run_path in full_actual_paths:
    summary_path = os.path.join(run_path, 'summary.json')
    config_path = os.path.join(run_path, 'config.yaml')

    if not os.path.exists(summary_path) or not os.path.exists(config_path):
        print(f'Skipping {run_path}: missing summary/config')
        continue

    with open(config_path, 'r') as f:
        config = yaml.safe_load(f)

    datasets = config.get('dataset', {}).get('datasets', [])
    if not datasets:
        print(f'Skipping {run_path}: no datasets block')
        continue

    train_case = None
    val_cases = []
    for ds in datasets:
        case_label = extract_case_label(ds.get('case_name', 'unknown'))
        if ds.get('split') == 'train' and train_case is None:
            train_case = case_label
        elif ds.get('split') == 'val':
            val_cases.append(case_label)

    if train_case is None:
        print(f'Skipping {run_path}: no train case found')
        continue

    with open(summary_path, 'r') as f:
        summary = json.load(f)

    val_metrics = summary.get('val')
    if not isinstance(val_metrics, list):
        print(f'Skipping {run_path}: unexpected val metrics')
        continue

    if len(val_metrics) != len(val_cases):
        print(f'Skipping {run_path}: mismatch val counts ({len(val_metrics)} != {len(val_cases)})')
        continue

    model_name = config.get('model', {}).get('name', 'unknown_model')
    key = (model_name, train_case)
    model_labels.setdefault(key, f"{model_name} / {train_case}")
    
    for case_label, metrics in zip(val_cases, val_metrics):
        if case_label not in target_cases:
            continue
        value = metrics.get(metric_name)
        if value is None:
            continue
        try:
            # print(key)
            # print(case_label)
            # print(value)
            results[key][case_label].append(float(value))
        except (TypeError, ValueError):
            print(f'Skipping value for {run_path} -> {case_label}: cannot convert {value}')
    
    # break

# print(results)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# gc_means: shape (n_groups, n_cats)
# gc_stds : shape (n_groups, n_cats)
# COLS    : list/array of category labels (length n_cats)

n_groups, n_cats = gc_means.shape
x = np.arange(n_groups)                          # group positions
bar_width = 0.8 / n_cats                         # width of each bar

fig, ax = plt.subplots(figsize=(14, 5.5))

# nicer default colors per category
cmap = plt.colormaps['tab10']

for i in range(n_cats):
    offset = (i - n_cats/2) * bar_width + bar_width/2
    y     = gc_means[:, i]
    yerr  = gc_stds[:, i]

    # If NaNs appear in means or stds, bar() will quietly skip drawing those bars
    ax.bar(
        x + offset, y,
        width=bar_width,
        yerr=yerr,
        label=f"Case {COLS[i]}",
        capsize=3,
        linewidth=0.8,
        edgecolor="black",
        alpha=0.95,
        color=cmap(i),
        error_kw=dict(elinewidth=1, capsize=3, capthick=1)
    )
    
title_size = 20
label_size = 16
tick_size = 14
legend_title_size = 14
legend_text_size = 12

# titles & labels
ax.set_title("GraphConv: Performance on different test cases across different train cases", pad=10, fontsize=title_size)
ax.set_ylabel("Mean PBL (±1 SD) (log)", fontsize=label_size)
ax.set_xlabel("Test Case", fontsize=label_size)

# x ticks
ax.set_xticks(x, [f"{COLS[j]}" for j in range(n_groups)])

# log y with helpful grid
ax.set_yscale('log')
ax.grid(True, which='both', axis='y', linestyle='--', linewidth=0.6, alpha=0.35)

# de-clutter spines
for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)

ax.legend(
    loc="upper right",
    title="Trained On",
    title_fontsize=legend_title_size,
    fontsize=legend_text_size
)

fig.tight_layout()
plt.show()
```
